```python
#Downloading Stopwords
nltk.download('stopwords')
```

```python
def process_text(text):

  #1 remove the punctuation
  #2 removing the stopwords
  #3 list of clean words

  #1
  nopun = [char for char in text if char not in string.punctuation]
  nopun = ''.join(nopun)

  #2
  clean_words = [word for word in nopun.split() if word.lower() not in stopwords.words('english')]

  #3
  return clean_words
```

```python
#showing the tokenisation
df['Email Text'].head(5).apply(process_text)
```

```python
df.dropna()
```

```python
#convert a collection of text to a matrix of tokens
from sklearn.feature_extraction.text import CountVectorizer
messages_bow = CountVectorizer(analyzer=process_text).fit_transform(df['Email Text'])
```

```python
#splitting the data for training and testing
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(messages_bow, df['Email Type'], test_size=0.20, random_state = 0)
```

```python
#Getting shape of messages_bow
messages_bow.shape
```

```python
#Create and train the classifier
from sklearn.naive_bayes import MultinomialNB
classifier = MultinomialNB().fit(X_train, y_train)
```

```python
#printing predictions
print(classifier.predict(X_train))

#actual
print(y_train.values)
```

```python
#evaluating the training set
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
pred = classifier.predict(X_train)
print(classification_report(y_train, pred))
print()
print('Confusion Matrix: \n', confusion_matrix(y_train, pred))
print()
print('Accuracy:', accuracy_score(y_train, pred))
```

```python
#printing predictions
print(classifier.predict(X_test))

#actual
print(y_test.values)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
pred = classifier.predict(X_test)
print(classification_report(y_test, pred))
print()
print('Confusion Matrix: \n', confusion_matrix(y_test, pred))
print()
print('Accuracy:', accuracy_score(y_test, pred))
```
